#### Soal 2.2 
Membuat dataframe yang menunjukkan data usia max, min & average, kemudian dikelompokkan berdasarkan profesi & gender

```python
df2 = df.groupby([df['occupation'],df['gender']]).describe()
df2 = df2['age'][['max','min','mean']]
df2 = df2.rename(
    columns={'max':'max_usia','min':'min_usia','mean':'rerata_usia'}
)
df2
```

#### Soal 2.3 
Membuat sebuah dataframe yang menunjukkan persentase pria & wanita tiap profesi

```python
gender = pd.crosstab(df.occupation, df.gender).apply(lambda i: i/i.sum(), axis= 1) *100
gender['%total'] = gender['F']+gender['M']

gender = gender.rename(columns={'F':'%female','M':'%male'})
gender = gender[['%male','%female','%total']]
gender
```
